```python
""" Для заданного пользователя (совпадает с вашим номером в таблице) 
рассчитать оценки для всех фильмов, которые он не оценил """
import numpy as np
import pandas as pd
import matplotlib
import math as m

#Параметры для расчета метрики сходства
my_user = 8
variant = my_user-1
kNN = 4

data = pd.read_csv('data.csv')
data = data.drop(data.columns[0],axis = 1)
data = data.replace({-1:None})
sred_v = round(data.iloc[[variant]].mean(axis = 1),3)
print("Средняя оценка", sred_v)


""" Составим метрику сходства для пользователей
    Для расчета метрики сходства был создан пустой масив Mat
    sum_p - числитель в выше рпедставленной формуле (сумма произведений оценок пользователей u, v)
    sum_u - левая часть произведения знаменателя - сумма квадратов оценок пользователя u
    sum_v - правая часть произведения знаменателя - сумма квадратов оценок пользователя v
"""
Mat = np.zeros((40,40)) 
data_to_eval = data.replace({None:0}).copy()
for user1 in range(40):
    for user2 in range(40):
        if(user1 != user2):
            sum_p = 0
            sum_u = 0
            sum_v = 0
            A = data_to_eval.iloc[lambda df: [user1,user2]].copy()
            for film in range(len(A.columns)):
                f_i = A[A.columns[film]].copy()
                if(int(f_i[user1]) != 0 and int(f_i[user2]) != 0):
                    sum_p = sum_p + int(f_i[user1]) * int(f_i[user2])
                    sum_u = sum_u + int(f_i[user1])**2 
                    sum_v = sum_v + int(f_i[user2])**2 
            sum_denumerator = m.sqrt(sum_u) * m.sqrt(sum_v)
            Mat[user1,user2] = round(sum_p / sum_denumerator,3)
print("Матрица сходства:\n",Mat)
print("Moй кортеж требуемых значений:\n", Mat[variant])
sim = Mat[variant]

"""Найдем индексы макисмальных значений (что и будет означать максимально похожих пользователей)
    Функция расставит в порядке возрастания, а я отслайсирую последние 4"""
ind = np.argpartition(Mat[variant], -kNN)[-kNN:]
print("Индексы максимально похожих", kNN,"пользователей:",ind) 

"""Фильмы, у которых нет оценок"""
data = pd.read_csv('data.csv')
data = data.drop(data.columns[0],axis = 1)
data_for_variant = data.replace({-1:0}).copy()
data_for_variant = data_for_variant.iloc[variant]
films_to_evaluate = dict()
sum_m = dict()
for i in range(30):
    if data_for_variant.iloc[i] == 0:
        films_to_evaluate[i] = 0 
        sum_m[i] = 0
print("Фильмы без оценок:",films_to_evaluate)

"""Для расчета оценки использовалась представленная выше формула.
   sum_m - сумма метрик сходства пользователей по каждрму фильму
   films_to_evaluate - словарь с индексами фильмов для оценки (не оценнеых прежде требуемым пользователем)
"""
sredv = sred_v[variant] 
data_to_get_sred = data.replace({-1:None}).copy()
for user1 in ind:
    dU = data_to_get_sred.iloc[[user1]] 
    sredu = round(dU.mean(axis = 1),3)
    sredu = sredu[user1] 
    for film in films_to_evaluate.keys():
        if(data.iloc[user1][film] != -1):
            sum_m[film] += sim[user1]
            films_to_evaluate[film] += sim[user1] * (data.iloc[user1][film] - sredu)
for i in films_to_evaluate.keys():
    films_to_evaluate[i] = round(sredv + round(films_to_evaluate[i] / abs(sum_m[i]),3),3)
print("Новые оценки для варианта:", films_to_evaluate)

names_for_films = data.iloc[variant].keys()
answer_21 = dict()
for i in films_to_evaluate.keys():
    answer_21[names_for_films[i]] = films_to_evaluate[i]
print("2.1 1 часть ответ: ",answer_21)
```

```python
import numpy as np
import pandas as pd
import matplotlib
import math as m
variant = 7
kNN = 4
days = [' Sat',' Sun']
place = ['h']
data = pd.read_csv('data.csv')
data = data.drop(data.columns[0],axis = 1)
contday = pd.read_csv("context_day.csv")
contday = contday.drop(contday.columns[0],axis = 1)
contpl = pd.read_csv("context_place.csv")
contpl = contpl.drop(contpl.columns[0],axis = 1)

close_ind = np.argpartition(Mat[variant], -kNN)[-kNN:]

duser = data.iloc[variant].copy()
duser = duser.transpose()
for i in range(30):
    if duser.iloc[i] != -1:
        duser.iloc[i] = np.NaN
answer = duser.dropna()
no_rate = answer.keys()

film_vac_ev = answer.copy()
data_to_get_sred = data.replace({-1:None}).copy()
films_to_eval_with_context = {}

for film in no_rate:
    sumt = 0
    sumb = 0
    for user in close_ind:
        dU = data_to_get_sred.iloc[[user]] 
        sredu = round(dU.mean(axis = 1),3)
        rate = data.loc[user, film]
        if rate != -1:
            if contday.loc[user, film] in days:
                sumb += sim[user]
                sumt += sim[user] * (data.loc[user, film] - sredu[user])
        films_to_eval_with_context[film] = round(sumt/sumb,3) if sumb !=0 else 0


for film in no_rate:
    sumt = 0
    sumb = 0
    for user in close_ind:
        dU = data_to_get_sred.iloc[[user]] 
        sredu = round(dU.mean(axis = 1),3)
        rate = data.loc[user, film]
        if rate != -1 and contpl.loc[user, film] == ' h':
            sumb += sim[user]
            sumt += sim[user] * (data.loc[user, film] - sredu[user])
            films_to_eval_with_context[film] += sumt/sumb if sumb !=0 else 0
            films_to_eval_with_context[film] = round(films_to_eval_with_context[film],3)


data_to_get_sred = data.replace({-1:None}).copy()
sred_v = round(data_to_get_sred.iloc[[variant]].mean(axis = 1),3)
sredv = sred_v[variant]


for film in no_rate:
    films_to_eval_with_context[film] = round(films_to_eval_with_context[film] + sredv,3)

c = 0
contday = contday.replace({' -1':np.NaN})
film_vac_ev = {}
for film in no_rate:
    c = contday[film].value_counts() 
    sSat = c[[' Sun']]
    sSun = c[[' Sat']]
    c = c.dropna()
    views = c.sum()
    """ основная формула расчёта: отношение суммы просмотров на выходных к общему количеству просмотров ( вычитается из 1, т.к. чем
    выше просмотры => тем меньше дробь, а я в дальнейшем планирую суммировать оценки => мне нужен не меньший, а больший рейтинг)"""
    sumVac = 1 - round((sSat[0] + sSun[0]) / views,3)
    film_vac_ev[film] = sumVac 
    
c = 0
film_home_ev = {}
contpl = contpl.replace({' -1':np.NaN})
for film in no_rate:
    c = contpl[film].value_counts() 
    sh = c[[' h']]
    c = c.dropna()
    views = c.sum()
    sumh = 1 - round(sh[0] / views,3)
    film_home_ev[film] = sumh
koef = {}
for film in no_rate:
    koef[film] = round(film_home_ev[film] * film_vac_ev[film],3)

eval_cont = {}
for film in no_rate:
    eval_cont[film] = round(films_to_eval_with_context[film] * koef[film],3)


max_key = max(eval_cont, key=lambda k: eval_cont[k])
print("Рекомендуменый фильм:",max_key, "\nЕго оценка:", eval_cont[max_key])
```

```python
from SPARQLWrapper import SPARQLWrapper, JSON
import pandas as pd
import requests
API_ENDPOINT = "https://www.wikidata.org/w/api.php"
params = {
    'action' : 'wbsearchentities',
    'format' : 'json',
    'language' : 'en',
    'search': movie_name
}
res = requests.get(API_ENDPOINT, params = params)
res.json()['search'][0]['description']
res.json()
```

```python
 filmcode=res.json()['search'][0]['id']
```

```python
sparql = SPARQLWrapper("https://query.wikidata.org/sparql")

sparql_query = """

SELECT ?filmLabels ?filmLabelsLabel
WHERE {
  BIND(wd:"""+filmcode+""" AS ?my_film)
  ?my_film wdt:P57 ?director.
  ?filmLabels wdt:P31 wd:Q11424.
  ?filmLabels wdt:P57 ?director.
  FILTER(?filmLabels != ?my_film).
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],ru". }
}
"""

sparql.setQuery(sparql_query)

sparql.setReturnFormat(JSON)
results = sparql.query().convert()
```
